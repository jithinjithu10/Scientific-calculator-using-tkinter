```python
import tkinter as tk
from math import *

# Create the main window
root = tk.Tk()
root.title("Scientific Calculator")
root.geometry("400x600")
root.resizable(False, False)

# Global variable to store the expression
expression = ""

# Functions to handle button press and calculation
def press(key):
    global expression
    expression += str(key)
    input_text.set(expression)

def clear():
    global expression
    expression = ""
    input_text.set("")

def delete():
    global expression
    expression = expression[:-1]
    input_text.set(expression)

def equal():
    try:
        global expression
        result = str(eval(expression))
        input_text.set(result)
        expression = result
    except:
        input_text.set("Error")
        expression = ""

# Entry widget for input
input_text = tk.StringVar()
input_frame = tk.Frame(root)
input_frame.pack()

input_field = tk.Entry(input_frame, textvariable=input_text, font=('Arial', 20), bd=10, insertwidth=4, width=16, justify='right')
input_field.grid(row=0, column=0, columnspan=4)

# Buttons for calculator
button_frame = tk.Frame(root)
button_frame.pack()

# Define buttons layout
buttons = [
    ['7', '8', '9', '/'],
    ['4', '5', '6', '*'],
    ['1', '2', '3', '-'],
    ['0', '.', '(', '+'],
    ['sqrt', 'pow', ')', 'C'],
    ['sin', 'cos', 'tan', 'del'],
    ['log', 'exp', 'pi', '=']
]

# Create buttons and position them on the grid
for i, row in enumerate(buttons):
    for j, button in enumerate(row):
        if button == '=':
            tk.Button(button_frame, text=button, width=5, height=2, font=('Arial', 15),
                      command=equal).grid(row=i, column=j, padx=5, pady=5)
        elif button == 'C':
            tk.Button(button_frame, text=button, width=5, height=2, font=('Arial', 15),
                      command=clear).grid(row=i, column=j, padx=5, pady=5)
        elif button == 'del':
            tk.Button(button_frame, text=button, width=5, height=2, font=('Arial', 15),
                      command=delete).grid(row=i, column=j, padx=5, pady=5)
        elif button == 'sqrt':
            tk.Button(button_frame, text=button, width=5, height=2, font=('Arial', 15),
                      command=lambda: press('sqrt(')).grid(row=i, column=j, padx=5, pady=5)
        elif button == 'pow':
            tk.Button(button_frame, text=button, width=5, height=2, font=('Arial', 15),
                      command=lambda: press('**')).grid(row=i, column=j, padx=5, pady=5)
        elif button in ('sin', 'cos', 'tan', 'log', 'exp'):
            tk.Button(button_frame, text=button, width=5, height=2, font=('Arial', 15),
                      command=lambda b=button: press(b + '(')).grid(row=i, column=j, padx=5, pady=5)
        elif button == 'pi':
            tk.Button(button_frame, text=button, width=5, height=2, font=('Arial', 15),
                      command=lambda: press('pi')).grid(row=i, column=j, padx=5, pady=5)
        else:
            tk.Button(button_frame, text=button, width=5, height=2, font=('Arial', 15),
                      command=lambda b=button: press(b)).grid(row=i, column=j, padx=5, pady=5)

# Start the application
root.mainloop()
```
